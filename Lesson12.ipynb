```python
students = pd.DataFrame({
   'name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Bob', 'Charlie'],
   'subject': ['Math', 'Math', 'Math', 'Science', 'Science', 'Science'],
   'score': [95, 78, 88, 92, 85, 90]
})
print(students)
```

```python
total_scores = students.groupby('name')['score'].sum()
print(total_scores)
```

```python
total_scores = students.groupby('name')['score'].mean()
print(total_scores)
```

```python
exam_count = students.groupby('name')['score'].count()
print(exam_count)
```

```python
sales_data = pd.DataFrame({
   'region': ['East', 'East', 'West', 'West', 'East', 'West'],
   'category': ['Electronics', 'Furniture', 'Electronics', 'Furniture', 'Electronics', 'Furniture'],
   'sales': [5000, 3000, 4500, 2800, 5200, 3100]
})
print(sales_data)
```

```python
regional_category = sales_data.groupby(['region', 'category'])['sales'].sum()
print(regional_category)
```

```python
regional_category = sales_data.groupby(['category', 'region'])['sales'].sum()
print(regional_category)
```

```python
result = sales_data.groupby(['region', 'category'])['sales'].sum().reset_index()
print(result)
```

```python
pivot = sales_data.pivot_table(
   values='sales',      # What to calculate
   index='region',      # Rows
   columns='category',  # Columns
   aggfunc='sum'        # How to aggregate
)
print(pivot)
```

```python
pivot = students.pivot_table(
   values='score',
   index='name',
   columns='subject',
   aggfunc='mean'
)
print(pivot)
```

```python
def score_range(series):
   return series.max() - series.min()


# Apply to grouped data
result = students.groupby('name')['score'].agg(score_range)
print(result)
```

```python
def score_range(x):
   return x.max() - x.min()


def above_90(x):
   return (x > 90).sum()  # Count scores above 90


result = students.groupby('name')['score'].agg(['mean', score_range, above_90])
print(result)
```

```python
# Real-world sales data
sales = pd.DataFrame({
   'date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02', '2024-01-03'],
   'salesperson': ['John', 'Mary', 'John', 'Mary', 'John'],
   'product': ['Laptop', 'Mouse', 'Keyboard', 'Laptop', 'Monitor'],
   'amount': [1200, 25, 75, 1150, 300]
})
```

```python
sales
```

```python
# Question 1: Who sold the most total?
top_seller = sales.groupby('salesperson')['amount'].sum().sort_values(ascending=False)
print("Top Sellers:")
print(top_seller)
```

```python
# Question 2: What's the average deal size per person?
avg_deal = sales.groupby('salesperson')['amount'].mean()
print("\nAverage Deal Size:")
print(avg_deal)
```

```python
# E-commerce transactions
orders = pd.DataFrame({
   'order_id': range(1, 13),
   'category': ['Electronics', 'Clothing', 'Electronics', 'Home'] * 3,
   'region': ['North', 'North', 'South', 'South'] * 3,
   'revenue': [1200, 300, 1500, 450, 1100, 280, 1600, 500, 1300, 320, 1400, 480],
   'quantity': [2, 5, 3, 2, 2, 6, 3, 2, 2, 4, 3, 2]
})
orders
```

```python
# Custom function: Calculate average price per item
def avg_price_per_item(group):
   return group['revenue'].sum() / group['quantity'].sum()
```

```python
# Best performing category
best_category = orders.groupby('category')['revenue'].sum().idxmax()
print(f"\nTop Category: {best_category}")
```
