```python
# Load data
df = pd.read_csv("students.csv")


# Quick overview
print("=== Student Report ===")
print(f"Total students: {len(df)}")
print(f"Average score:  {df['Score'].mean():.1f}")
print(f"Highest score:  {df['Score'].max()}")
print(f"Lowest score:   {df['Score'].min()}")
```

```python
# Find top performers (score above average)
avg = df["Score"].mean()
top_students = df[df["Score"] > avg]


print(f"Class average: {avg:.1f}")
print(f"\nTop performers ({len(top_students)} students):")
print(top_students[["Name", "Score", "Grade"]])

```

```python
# Find the best student
best = df[df["Score"] == df["Score"].max()]
print(f"\nBest student:")
print(best[["Name", "Score"]])
```

```python
df["Name", "Score"]     # ❌ KeyError
df[["Name", "Score"]]   # ✅ Correct
```

```python
df[df["Score"] > 80 and df["Age"] < 22]    # ❌ ValueError
df[(df["Score"] > 80) & (df["Age"] < 22)]  # ✅ Correct
```

```python
df[df["Score"] > 80 & df["Age"] < 22]      # ❌ Wrong result (operator precedence)
df[(df["Score"] > 80) & (df["Age"] < 22)]  # ✅ Correct
```

```python
df[df["Grade"] = "A"]    # ❌ SyntaxError — this is assignment
df[df["Grade"] == "A"]   # ✅ Correct — this is comparison
```
